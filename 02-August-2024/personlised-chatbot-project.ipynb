```python
import requests
from PyQt5 import QtWidgets, QtGui, QtCore, QtGui

API_URL = "https://api-inference.huggingface.co/models/google-bert/bert-large-uncased-whole-word-masking-finetuned-squad"
headers = {"Authorization": "Bearer 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"}

def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()

class FadeInEffect(QtCore.QPropertyAnimation):
    def __init__(self, widget, duration=1000, **kwargs):
        super().__init__(widget, **kwargs)
        self.setDuration(duration)
        self.setPropertyName(b"windowOpacity")
        self.setStartValue(0)
        self.setEndValue(1)
        self.setEasingCurve(QtCore.QEasingCurve.InOutQuad)

class BertQAApp(QtWidgets.QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        # Set up the main layout
        self.setWindowTitle('BERT Question Answering')
        self.setGeometry(100, 100, 800, 600)
        self.setStyleSheet("background-color: #f4f4f4;")

        # Main layout
        layout = QtWidgets.QVBoxLayout()
        layout.setSpacing(20)
        layout.setContentsMargins(20, 20, 20, 20)

        # Title label
        title = QtWidgets.QLabel('BERT Question Answering')
        title.setFont(QtGui.QFont('Arial', 18, QtGui.QFont.Bold))
        title.setAlignment(QtCore.Qt.AlignCenter)
        title.setStyleSheet("color: #333;")
        layout.addWidget(title)

        # Question input
        question_label = QtWidgets.QLabel('Question:')
        question_label.setFont(QtGui.QFont('Arial', 12))
        layout.addWidget(question_label)

        self.question_entry = QtWidgets.QLineEdit()
        self.question_entry.setFont(QtGui.QFont('Arial', 12))
        self.question_entry.setStyleSheet("border: 1px solid #ddd; border-radius: 5px; padding: 5px;")
        layout.addWidget(self.question_entry)

        # Context input
        context_label = QtWidgets.QLabel('Context:')
        context_label.setFont(QtGui.QFont('Arial', 12))
        layout.addWidget(context_label)

        self.context_entry = QtWidgets.QTextEdit()
        self.context_entry.setFont(QtGui.QFont('Arial', 12))
        self.context_entry.setStyleSheet("border: 1px solid #ddd; border-radius: 5px; padding: 5px;")
        layout.addWidget(self.context_entry)

        # Submit button
        submit_button = QtWidgets.QPushButton('Get Answer')
        submit_button.setFont(QtGui.QFont('Arial', 12))
        submit_button.setStyleSheet("""
            background-color: #4CAF50;
            color: white;
            border: none;
            border-radius: 5px;
            padding: 10px;
            font-size: 14px;
        """)
        submit_button.clicked.connect(self.get_answer)
        layout.addWidget(submit_button)

        # Answer label
        self.answer_label = QtWidgets.QLabel('Answer: ')
        self.answer_label.setFont(QtGui.QFont('Arial', 12))
        self.answer_label.setWordWrap(True)
        self.answer_label.setStyleSheet("color: #333;")
        layout.addWidget(self.answer_label)

        # Set the layout to the main window
        self.setLayout(layout)

        # Apply fade-in effect to the main widget
        self.fade_in_effect = FadeInEffect(self)
        self.fade_in_effect.start()

    def get_answer(self):
        question = self.question_entry.text()
        context = self.context_entry.toPlainText().strip()

        if question and context:
            output = query({
                "inputs": {
                    "question": question,
                    "context": context
                },
            })
            if "error" in output:
                self.answer_label.setText("Error: " + output["error"])
                self.answer_label.setStyleSheet("color: red;")
            else:
                answer = output.get('answer', 'No answer found')
                self.answer_label.setText("Answer: " + answer)
                self.answer_label.setStyleSheet("color: black;")
        else:
            self.answer_label.setText("Please enter both question and context.")
            self.answer_label.setStyleSheet("color: red;")

if __name__ == '__main__':
    import sys
    app = QtWidgets.QApplication(sys.argv)
    ex = BertQAApp()
    ex.show()
    sys.exit(app.exec_())
```
