```python
reshaped_data = df.melt(id_vars=['Name', 'Age', 'Gender'],var_name='Subject', value_name='Score')

reshaped_data
```

```python
def Calculate_grade(Total):
   if (270<Total<=300):
      return 'A'
   elif (255<=Total<270):
      return 'B'
   elif Total<105:
      return 'Fail'
   else:
      return 'Incorrect'
   
```
