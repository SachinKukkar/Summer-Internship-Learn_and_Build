2D Plot : 
 
 It is used for 2 columns
 Both the data need to be numerical or 
 One is numerical and other is categorical

Use case : 
  
  Time series data like stock , covid cases , etc.
  

```python
price = [4892,5492,2304,7403,3058]
year = [2015,2016,2017,2018,2019]
```

```python
plt.plot(year,price)
plt.show()
```

```python
import matplotlib.pyplot as plt

batsmen = ['Sachin', 'Rohit', 'Rahane', 'Pant', 'Dhawan']
runs = [1200, 1100, 900, 950, 850]

plt.plot(batsmen,runs)

plt.show()
```

```python
runs_2 = [10000,21000,4539,6482,1283]

batsmen_2 = ['Sachin','MSD','Rohit','Dhawan','Rahane']

plt.plot(batsmen_2,runs_2,label="batsmen runs") # Tells what graph line is indicating

plt.xlabel("Batsmen Name") # Labels the X-axis

plt.ylabel("Runs scored")  # Lables the Y-axis

plt.legend()    # For printing the legend , i.e name of the graph line

plt.show()   
```

Bar Chart : 
  Bivariate analysis : Numerical v/S Categorical values

  Use Case : 

  Aggregate analysis of group 

```python
children = [10,20,40,10,30]

color = ['red','blue','green','yellow' , 'black']

ax = plt.axes()

plt.bar(color,children,color='black')

ax.set_facecolor("red")

plt.show()
```
